### EmployeeSQL Visualization

```python
#Dependencies
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from sql_config import password
import pandas as pd
import statistics as stat
```

```python
#Create Engine and Connection
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/sql-challenge')
connection = engine.connect()
```

```python
#Extract data from table
data = pd.read_sql('SELECT * FROM salaries',connection)
```

```python
data
```

```python
print(max(data.salary))
```

```python
#Creates histogram using specified bins
fig, ax = plt.subplots()
ax.hist(data['salary'],bins = [39999,49999,59999,69999,79999,89999,99999,109999,119999,129999],color='maroon')
plt.title('Distribution of Salaries for Employees')
plt.ylabel('Number of Employees')
plt.xlabel('Salary ($)')

ax.xaxis.set_ticks([45000,55000,65000,75000,85000,95000,105000,115000,125000])
                   
ax.set_xticklabels(['$40k-$50k','$50k-$60k','$60k-$70k','$70k-$80k','$80k-$90k','$90k-$100k','$100k-$110k','$110k-$120k','$120k-$130k'])
plt.xticks(rotation=45)
plt.show()
```

```python
data_2 = pd.read_sql('select s.salary, t.title, t.title_id' \
                     + ' from salaries s, titles t, employees e' \
                     + ' where e.emp_title_id = t.title_id' \
                     +  ' and s.emp_no = e.emp_no',connection)

eng = data_2.loc[data_2['title'] == 'Engineer']
eng_avg_sal = round(stat.mean(eng['salary']),0)

seng = data_2.loc[data_2['title'] == 'Senior Engineer']
seng_avg_sal = round(stat.mean(seng['salary']),0)

assist_eng = data_2.loc[data_2['title'] == 'Assistant Engineer']
a_eng_avg_sal = round(stat.mean(assist_eng['salary']),0)

staff = data_2.loc[data_2['title'] == 'Staff']
staff_avg_sal = round(stat.mean(staff['salary']),0)

sen_staff = data_2.loc[data_2['title'] == 'Senior Staff']
sen_staff_avg_sal = round(stat.mean(sen_staff['salary']),0)

tech_lead = data_2.loc[data_2['title'] == 'Technique Leader']
tech_lead_avg_sal = round(stat.mean(tech_lead['salary']),0)

manage = data_2.loc[data_2['title'] == 'Manager']
manage_avg_sal = round(stat.mean(manage['salary']),0)

avg_sal = [staff_avg_sal, sen_staff_avg_sal, a_eng_avg_sal, \
          eng_avg_sal, seng_avg_sal, tech_lead_avg_sal,\
          manage_avg_sal]

fig, ax = plt.subplots()
ax.bar(['Staff','Senior Staff','Assistant Engineer',\
       'Enginer','Senior Engineer','Technique Leader','Manager'],\
       avg_sal, color = 'maroon')
plt.title('Average Salary by Title')
plt.ylabel('Average Salary ($)')
plt.xticks(rotation=30)
plt.show()
```
